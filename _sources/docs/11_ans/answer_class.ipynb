## 演習1：クラスの作成

## 演習3：インスタンスメソッド
演習2で作成したクラスを再定義して，以下の情報を出力するインスタンスメソッド `print_data` を作成せよ．

```
マリー・キュリーの会員IDはMK210です
アルベルト・アインシュタインの会員IDはAE314です
ニコラ・テスラの会員IDはNT271です
```


````python
# 解答

class Member:
    def __init__(self, nickname: str, account_id: str):
        self.nickname = nickname
        self.account_id = account_id

    def print_data(self) -> None:
        print(f'{self.nickname}の会員IDは{self.account_id}です')

club_members = [
    Member('マリー・キュリー', 'MK210'),
    Member('アルベルト・アインシュタイン', 'AE314'),
    Member('ニコラ・テスラ', 'NT271'),
]

for member in club_members:
    member.print_data()
````

## 演習4：クラスのまとめ
講義資料に含まれていないあなた自身が考えた独自の機能を持つクラスを定義して，インスタンスメソッドから実装した処理の呼び出しを行うコードを作成せよ．独自であれば**どのような機能でも良い**．また **説明** と書かれたMarkdownセルに実装したクラスの機能について説明せよ．

一からクラスを作成することが難しければ演習1-3のクラスをベースにしても良い．

````python
# 解答例

class ReadingMember:
    def __init__(self, nickname: str, account_id: str, monthly_goal: int):
        self.nickname = nickname
        self.account_id = account_id
        self.monthly_goal = monthly_goal
        self.books_read = 0

    def log_books(self, count: int) -> None:
        self.books_read = count + self.books_read

    def progress_label(self) -> str:
        ratio = self.books_read / self.monthly_goal
        if ratio >= 1:
            return '目標達成'
        if ratio >= 0.5:
            return '折り返し地点'
        return 'これから'

    def print_progress(self) -> None:
        label = self.progress_label()
        print(f'{self.nickname}（ID:{self.account_id}）は{self.books_read}冊読了：{label}')

curie = ReadingMember('マリー・キュリー', 'MK210', monthly_goal=5)
einstein = ReadingMember('アルベルト・アインシュタイン', 'AE314', monthly_goal=4)
tesla = ReadingMember('ニコラ・テスラ', 'NT271', monthly_goal=6)

curie.log_books(4)
einstein.log_books(1)
tesla.log_books(6)

for reading_member in (curie, einstein, tesla):
    reading_member.print_progress()
````

**説明**

サークル内の読書目標を可視化できるクラス `ReadingMember` を実装した．
- ニックネーム，ID，月間読書目標冊数を保持する．
- `log_books` で読了冊数を蓄積し，`progress_label` で達成状況を判定する．
- `print_progress` を呼び出すと現在の読了冊数と達成ラベルをまとめて表示できる．
メソッドを呼び出すだけでその月の進捗を共有できるため，演習1-3で扱った会員情報の応用例になる．
