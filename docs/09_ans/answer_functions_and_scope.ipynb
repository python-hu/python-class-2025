### 演習3-1

**関数名：** `get_odd_numbers`  
**引数：** `numbers`（整数が格納されたリスト）  
**処理の内容：** 引数で渡された整数のリスト（`numbers`）の中に，奇数が何個含まれているかカウントし，そのカウント数を返す．  
**例：** `get_odd_numbers([1, 2, 3, 4, 5])` とすると，値 `3` を返すようにする．

````python
# 解答例1
def get_odd_numbers(numbers):
    count = 0
    for num in numbers:
        if num % 2 != 0:  # 奇数の場合
            count += 1
    return count

result = get_odd_numbers([1, 2, 3, 4, 5])
print(result)
````

````python
# 解答例2
def get_odd_numbers(numbers):
    return len([num for num in numbers if num % 2 != 0])

result = get_odd_numbers([1, 2, 3, 4, 5])
print(result)
````

### 演習3-2

**関数名：** `get_divisors`  
**引数：** `number`  
**処理の内容：** 引数で渡された正の整数 `number` の約数をすべて求め，リストとして返す．  
**例：** `get_divisors(12)` とすると，リスト `[1, 2, 3, 4, 6, 12]` を返す．  
**ヒント：** 1から `number` までの整数それぞれで `number` を割り，余りが0になるものが約数である．

````python
# 解答例
def get_divisors(number):
    divisors = []
    for i in range(1, number + 1):
        if number % i == 0:  # iがnumberの約数の場合
            divisors.append(i)
    return divisors

result = get_divisors(12)
print(result)
````

### 演習3-3

**関数名：** `greet_with_time`  
**引数：** `name`, `time="morning"`  
**処理の内容：** 引数 `name` に指定された名前と，`time` に指定された時間帯に応じた挨拶メッセージを表示する．`time` には `"morning"`, `"afternoon"`, `"evening"` のいずれかを指定できる．`time` はデフォルト引数として定義し，呼び出し時に値が指定されない場合は `"morning"` が設定されるようにせよ．

**実行・出力例1**

```python
greet_with_time("Alice", "afternoon")
```

```:python
Good afternoon, Alice!
```

**実行・出力例2**（`time` に値を指定しない場合，`time="morning"` として扱われる）

```python
greet_with_time("Bob")
```

```:python
Good morning, Bob!
``` 

**実行・出力例3**

```python
greet_with_time("Charlie", "evening")
```

```:python
Good evening, Charlie!
``` 

````python
# 解答例1
def greet_with_time(name, time="morning"):
    if time == "morning":
        print(f"Good morning, {name}!")
    elif time == "afternoon":
        print(f"Good afternoon, {name}!")
    elif time == "evening":
        print(f"Good evening, {name}!")
    else:
        print(f"Hello, {name}!")

# テスト
greet_with_time("Alice", "afternoon")
greet_with_time("Bob")
greet_with_time("Charlie", "evening")
````

````python
# 解答例2
def greet_with_time(name, time="morning"):
    greetings = {
        "morning": "Good morning",
        "afternoon": "Good afternoon",
        "evening": "Good evening"
    }
    message = greetings.get(time, "Hello")
    print(f"{message}, {name}!")

# テスト
greet_with_time("Alice", "afternoon")
greet_with_time("Bob")
greet_with_time("Charlie", "evening")
````

### 演習3-4

**関数名：** `calculate_product`  
**引数：** `*numbers`  
**処理の内容：** 可変長引数で渡された複数の数値の積を返す．  
**例：** `calculate_product(2, 3, 4, 5)` とすると，値 `120` が返される．  
**ヒント：** 積を求める場合は，初期値を1にして，各数値を順番に掛け算していく．

````python
# 解答例
def calculate_product(*numbers):
    product = 1
    for num in numbers:
        product *= num
    return product

result = calculate_product(2, 3, 4, 5)
print(result)
````

### 演習3-5

**関数名：** `apply_operation`  
**引数：** `numbers`（数値が格納されたリスト）, `operation`（関数）  
**処理の内容：** 第1引数として与えられたリストの各要素に対して，第2引数として与えられた関数を適用し，結果をリストとして返す．  
第2引数に指定する関数としては，以下の `square` 関数を使用せよ．
```python
# 受け取った数値の2乗を返す
def square(value):
    return value ** 2
```

**実行・出力例**
```python
numbers = [1, 2, 3, 4, 5]

result = apply_operation(numbers, square)
print(result)
```

```:python
[1, 4, 9, 16, 25]
```

````python
# 解答例
def square(value):
    return value ** 2

def apply_operation(numbers, operation):
    result = []
    for num in numbers:
        result.append(operation(num))
    return result

numbers = [1, 2, 3, 4, 5]
result = apply_operation(numbers, square)
print(result)
````

### 演習3-6

先ほどの演習3-5の課題について， `square` 関数を定義せずに，ラムダ式を用いて同様の処理を実現せよ．

````python
# 解答例
def apply_operation(numbers, operation):
    result = []
    for num in numbers:
        result.append(operation(num))
    return result

numbers = [1, 2, 3, 4, 5]
result = apply_operation(numbers, lambda x: x ** 2)
print(result)
````
